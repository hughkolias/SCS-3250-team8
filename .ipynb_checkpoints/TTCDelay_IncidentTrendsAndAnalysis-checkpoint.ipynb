```python
# Import packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from statsmodels.tsa.seasonal import seasonal_decompose
from pylab import rcParams
import seaborn as sns
%matplotlib inline
sns.set()

%matplotlib inline
```

```python
# Import datasets
delay = pd.read_csv('data/all_years.csv', delimiter='\t')
codes_srt = pd.read_csv('data/codes_srt.csv', delimiter='\t')
codes_sub = pd.read_csv('data/codes_sub.csv', delimiter='\t')
station_names = pd.read_csv('data/station_names.csv')
codes_bins = pd.read_csv('data/codes_bins.csv')

# Merge Datasets
delay = pd.merge(left=delay, right=codes_srt, how='left', left_on=delay['Code'], 
        right_on=codes_srt['SRT RMENU CODE']).drop(columns='key_0',axis=1)

delay = pd.merge(left=delay, right=codes_sub, how='left', 
                left_on=delay['Code'], right_on=codes_sub['SUB RMENU CODE']).drop(columns='key_0', axis=1)

delay = pd.merge(left=delay, right=station_names, how='left',
                left_on=delay['Station'], right_on=station_names['data_station']).drop(columns=['Unnamed: 0','key_0','data_station'],axis=1)

# Create Combo Code Column
delay['combo code'] = delay['CODE DESCRIPTION_y'].fillna(delay['CODE DESCRIPTION_y'])

delay = pd.merge(left=delay, right=codes_bins, how='left', left_on=delay['combo code'], 
                right_on=codes_bins['combo code']).drop(columns=['key_0','combo code_x'], axis=1)
delay = delay.rename(columns={'combo code_y': 'combo code'})
```

```python
pd.merge(left=delay, right=codes_srt, how='left', left_on=delay['Code'], 
        right_on=codes_srt['SRT RMENU CODE']).drop(columns='key_0',axis=1)

```

```python
# Filter Dataset
delay = delay.drop(columns=['SRT RMENU CODE','CODE DESCRIPTION_x','SUB RMENU CODE', 
                   'CODE DESCRIPTION_y'])
```

## Missing Values

```python
delay.isnull().mean().sort_values().plot(kind='barh')
```

### Bound
- Directional in one direction, nulls can be expected as some delays are when subways are not in transit
- Complete Nulls with 'NA'

### real_station
- some "Station" in original datasets are not real subway stations, therefore we can fill Nulls with 'Misc'

### combo code
- some delays are not reported with a reason
- we cant complete this data
- fill in all Nulls with 'misc'

```python
# Complete Missing Values if possible
delay['Bound'] = delay['Bound'].fillna('NA')
delay['real_station'] = delay['real_station'].fillna('Misc')
delay['combo code'] = delay['combo code'].fillna('Misc')
```

## Data Cleaning

```python
# What are the Data Types are we working with?
delay.info()
```

```python
# Convert data to Datetime
delay['Date'] = pd.to_datetime(delay['Date'])

# Generate Month/Year Variables
delay['Month'] = [i.month for i in delay['Date']]
delay['Year'] = [i.year for i in delay['Date']]
delay['Day'] = [i.dayofweek for i in delay['Date']]
```

```python
# Get Hours columns for Min Delay/Gap
delay['Min Delay Hours'] = delay['Min Delay']/60
delay['Min Gap Hours'] = delay['Min Gap']/60
```

```python
# Generate Time Category
delay['Time'] = pd.to_datetime(delay['Time'])
```

```python
def hour_of_day(time):
    return time.hour

def time_classifier(time):
    hour = hour_of_day(time)
    if (hour >= 5 and hour < 12):
        return 'Morning'
    elif (hour >= 12 and hour < 17):
        return 'Afternoon'
    elif (hour >= 17 and hour < 23):
        return 'Evening'
    elif (hour >= 23 or (hour >= 0 and hour < 5)):
        return 'Hooligan Hour'

# Generate Time Groupings
delay['Hour'] = delay['Time'].apply(hour_of_day)
delay['time_of_day'] = delay['Time'].apply(time_classifier)
```

```python
monthly_ridership = pd.read_csv('data/monthly_ridership.csv')
monthly_ridership['date'] = pd.to_datetime(monthly_ridership['Year'].astype(str) + monthly_ridership['Period'], format='%Y%b')
monthly_ridership = monthly_ridership[(monthly_ridership['date'] >  pd.to_datetime('2013-12-01'))]
monthly_ridership.set_index(['date'],inplace=True)
    
summary_df = delay.copy()
summary_df['count'] = summary_df.groupby(['Date'])['Date'].transform('count')
summary_df = summary_df.drop_duplicates(subset=['Date']).set_index('Date')
summary_df['ridership'] = monthly_ridership[['Value']]
time_series_df = summary_df.resample('MS')['count'].sum()
ridership_series_df = summary_df.resample('MS')['ridership'].sum()

normalized = time_series_df / (ridership_series_df / 100000)

rcParams['figure.figsize'] = 11, 9

decomposition = sm.tsa.seasonal_decompose(normalized, model='additive')
fig = decomposition.plot()
plt.set_cmap('plasma')
plt.suptitle(t='TTC Monthly Incidents', y=1.02, fontsize=14)
plt.show()
```

```python
summary_df = delay.copy()

#
monthly_ridership = pd.read_csv('data/monthly_ridership.csv')
monthly_ridership['date'] = pd.to_datetime(monthly_ridership['Year'].astype(str) + monthly_ridership['Period'], format='%Y%b')
monthly_ridership = monthly_ridership[(monthly_ridership['date'] >  pd.to_datetime('2013-12-01'))]
monthly_ridership.set_index(['date'],inplace=True)

# count for frequency per hour

summary_df['count'] = summary_df.groupby('Date')['Date'].transform('count')
summary_df = summary_df.drop_duplicates(subset=['Date']).set_index('Date').fillna(summary_df.bfill())
summary_df['ridership'] = monthly_ridership[['Value']]

time_series_df = summary_df.resample('MS')['count'].sum()
ridership_series_df = monthly_ridership.resample('MS')['Value'].sum()

normalized = time_series_df / (ridership_series_df / 1000000)

normalized = pd.DataFrame({'date':normalized.index, 'incident_count':normalized.values}).dropna()
normalized['month'] = [i.month for i in normalized['date']]


normalized['count'] = normalized.groupby('month')['incident_count'].sum()
normalized = normalized.dropna().sort_values('month')
plot = normalized.set_index('month')[['incident_count']].plot(kind='bar', colormap='plasma', title="Total Monthly Incidents (per 100k pax)", figsize=(10,5))
plot.axhline(y=normalized['incident_count'].mean())
plot.set_xticklabels(['Jan', 'Feb', 'March', 'Apr', 'May', 'June', 'July', 'Aug', 'Sept', 'Oct', 'Nov', 'Dec'])
```

```python
summary_df = delay.copy()

summary_df['count'] = summary_df.groupby(['Day'])['Day'].transform('count')

summary_df = summary_df.drop_duplicates(subset=['Day']).sort_values(['Day'])
data_frame = summary_df.plot(kind='bar', x='Day', y='count', colormap='plasma', title="Daily Frequency of All Incidents", figsize=(10,5))
data_frame.axhline(y=summary_df['count'].mean())
data_frame.set_ylabel("Frequency",fontsize=12)
data_frame.set_xticklabels(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
```

```python
for code_bin in unique_code_bins:
    # filter out by code_bin
    sorted_by_code_bin = delay[delay['code bin']==code_bin]
    # count for frequency per hour
    sorted_by_code_bin['count'] = sorted_by_code_bin.groupby('Hour')['Hour'].transform('count')
    sorted_df = sorted_by_code_bin[['code bin', 'Hour', 'count']].drop_duplicates(subset=['code bin', 'Hour']).sort_values(['code bin', 'Hour'])
    # fill out hour
    new_index = range(0,24)
    cleaned_df = sorted_df.set_index("Hour").reindex(new_index).reset_index()
    cleaned_df['code bin'] = cleaned_df['code bin'].fillna(unique_code_bins[3])
    cleaned_df['count'] = cleaned_df['count'].fillna(0)
    
    #plot
    plot = cleaned_df.plot(kind='bar', x='Hour', y='count', colormap='plasma', title=code_bin, figsize=(10,5))
    plot.set_ylabel("Frequency",fontsize=12)
    plot.axhline(y=cleaned_df['count'].mean())
```

```python
# Monthly trend analysis per incident category

for code_bin in unique_code_bins:
    #
    monthly_ridership = pd.read_csv('data/monthly_ridership.csv')
    monthly_ridership['date'] = pd.to_datetime(monthly_ridership['Year'].astype(str) + monthly_ridership['Period'], format='%Y%b')
    monthly_ridership = monthly_ridership[(monthly_ridership['date'] >  pd.to_datetime('2014-02-01'))]
    monthly_ridership.set_index(['date'],inplace=True)

    # filter out by code_bin
    summary_df = delay[delay['code bin']==code_bin].dropna()
    # count for frequency per hour

    summary_df['count'] = summary_df.groupby('Date')['Date'].transform('count')
    summary_df = summary_df.drop_duplicates(subset=['Date']).set_index('Date').fillna(summary_df.bfill())
    summary_df['ridership'] = monthly_ridership[['Value']]
    
    time_series_df = summary_df.resample('MS')['count'].sum()
    ridership_series_df = monthly_ridership.resample('MS')['Value'].sum()

    normalized = time_series_df / (ridership_series_df / 1000000)
    
    rcParams['figure.figsize'] = 11, 9

    decomposition = sm.tsa.seasonal_decompose(normalized.dropna(), model='additive')
    fig = decomposition.plot()
    plt.set_cmap('plasma')
    plt.suptitle(t=code_bin + ' (per 100k pax)', y=1.02, fontsize=14)
    plt.show()





```

```python
# Total Number of incidents per day

for code_bin in unique_code_bins:
    # filter out by code_bin
    sorted_by_code_bin = delay[delay['code bin']==code_bin]
    # count for frequency per hour
    sorted_by_code_bin['count'] = sorted_by_code_bin.groupby('Day')['Day'].transform('count')
    sorted_df = sorted_by_code_bin[['code bin', 'Day', 'count']].drop_duplicates(subset=['code bin', 'Day']).sort_values(['code bin', 'Day'])
    # fill out hour
    new_index = range(0,7)
    cleaned_df = sorted_df.set_index("Day").reindex(new_index).reset_index()
    cleaned_df['code bin'] = cleaned_df['code bin'].fillna(unique_code_bins[3])
    cleaned_df['count'] = cleaned_df['count'].fillna(0)
    
    #plot
    plot = cleaned_df.plot(kind='bar', x='Day', y='count', colormap='plasma', title=code_bin, figsize=(10,5))
    plot.set_ylabel("Frequency",fontsize=12)
    plot.axhline(y=cleaned_df['count'].mean())
    plot.set_xticklabels(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])

```
